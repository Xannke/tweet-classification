```python
def generate_text_preprocessing(data_loader, preprocess: Preprocess):
    data = data_loader.load_tweet(data_loader.train_path)
        
    vocab = data['turn1'].apply(preprocess.text_standardization)\
            .append(data['turn2'].apply(preprocess.text_standardization))\
            .append(data['turn3'].apply(preprocess.text_standardization))\

    text_processing = TextToVec()
    text_processing.fit(data_loader, vocab, 300)
    text_processing.save("tp-all-300.save")
```

```python
generate_text_preprocessing(data_loader, preprocess)
```
